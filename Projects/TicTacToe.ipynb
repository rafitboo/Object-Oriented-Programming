```python
#tictactoe game
class player:
    def __init__(self, name, sign):
        self.name = name
        self.sign = sign
        self.wins = 0
        self.played = 0
        self.win_status = False


class Board:
    def __init__(self):
        self.board = [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "]

    def display(self):

        print(f" {self.board[1]} | {self.board[2]} | {self.board[3]} ")
        print("-----------")
        print(f" {self.board[4]} | {self.board[5]} | {self.board[6]} ")
        print("-----------")
        print(f" {self.board[7]} | {self.board[8]} | {self.board[9]} ")

    def play(self):
        inp = input("Do you want to play?(y/n)\n")
        if inp.lower() == "y":
            name = input("Enter player 1 name : ")
            # sign=input("Enter player 1 sign : ")
            while True:
                sign = input("Enter player 1 sign : ").upper()
                if sign.upper() == "X" or sign == "0":
                    break
                else:
                    print("Invalid Sign. Please input again.")
                    continue
            p1 = player(name, sign)
            if p1.sign.upper() == "X":
                name = input("Enter player 2 name : ")
                p2 = player(name, "0")
            elif p1.sign.upper() == "0":
                name = input("Enter player 2 name : ")
                p2 = player(name, "X")
            print(f"Welcome {p1.name}, your sign is {p1.sign} and {p2.name}, your sign is {p2.sign}.")
            self.start(p1, p2)
        elif inp.lower() == "n":
            print("See You Again")
        else:
            print("Invalid input")
            self.play()

    def start(self, p1, p2):
        self.display()
        i = 1
        while i <= 9:
            if i % 2 == 1:
                print("Player 1's turn :")
                var = int(input("Choose a number(1-9):\n"))
                if var > 9 or var < 0:
                    print("Invalid Input")
                    self.display()
                elif self.board[var] == "X" or self.board[var] == "0":
                    print("The box is taken already!")
                    self.display()
                else:
                    self.board[var] = p1.sign
                    self.display()
                    i += 1

            elif i % 2 == 0:
                print("Player 2's turn :")
                var = int(input("Choose a number(1-9):\n"))
                if var > 9 or var < 1:
                    print("Invalid Input")
                    self.display()
                elif self.board[var] == "X" or self.board[var] == "0":
                    print("The box is taken already!")
                    self.display()
                else:
                    self.board[var] = p2.sign
                    self.display()
                    i += 1
            if i >= 5:
                if self.check_winner(p1, p2):
                  self.board = [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "]
                  self.play()
                  break
                elif i == 10 and p1.win_status == False and p2.win_status == False:
                    print("It's a draw :(")
                    self.board = [" ", " ", " ", " ", " ", " ", " ", " ", " ", " "]
                    self.play()
                    break
                else:
                    continue

    def check_winner(self, p1, p2):
        for i in range(1, 9, 3):
            # check rows
            if self.board[i] == self.board[i + 1] == self.board[i + 2] and self.board[i] != " ":
                if self.board[i] == p1.sign:
                    print(f"{p1.name} wins!")
                    p1.wins += 1
                    p1.win_status = True
                    return True
                else:
                    print(f"{p2.name} wins!")
                    p2.wins += 1
                    p2.win_status = True
                    return True

        for i in range(1, 4):
            # check columns
            if self.board[i] == self.board[i + 3] == self.board[i + 6] and self.board[i] != " ":
                if self.board[i] == p1.sign:
                    print(f"{p1.name} wins!")
                    p1.wins += 1
                    p1.win_status = True
                    return True

                else:
                    print(f"{p2.name} wins!")
                    p2.wins += 1
                    p2.win_status = True
                    return True

                # check diagonals
        if self.board[1] == self.board[5] == self.board[9] and self.board[1]!=" ":
            if self.board[1]==p1.sign:
                print(f"{p1.name} wins!")
                p1.wins += 1
                p1.win_status = True
                return True
            else:
                print(f"{p2.name} wins!")
                p2.wins += 1
                p2.win_status = True
                return True

        elif self.board[3] == self.board[5] == self.board[7] and self.board[3]!=" ":
            if self.board[3]==p1.sign:
                print(f"{p1.name} wins!")
                p1.wins += 1
                p1.win_status = True
                return True
            else:
                print(f"{p2.name} wins!")
                p2.wins += 1
                p2.win_status = True
                return True

        else:
            return False


Board().play()
```
