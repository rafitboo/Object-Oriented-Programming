```python
#TASK 5
var="OhMyBRCU@"
lower=False
upper=False
dig=False
special=False
for i in var:
  if 65<=ord(i)<=90:
    upper=True
  elif 97<=ord(i)<=122:
    lower=True
  elif 48<=ord(i)<=57:
    dig=True
  elif i=="_" or i=="$" or i=="#" or i=="@":
    special=True
out=""
if lower==False:
  if out!="":
    out+=", Lowercase missing"
  else:
    out+="Lowercase missing"
if upper==False:
  if out!="":
    out+=", Uppercase missing"
  else:
    out+="Uppercase missing"
if dig==False:
  if out!="":
    out+=", Digit missing"
  else:
    out+="Digit missing"
if special==False:
  if out!="":
    out+=", Special character missing"
  else:
    out+="Special character missing"

if lower==True and upper==True and dig==True and special==True :
  print("OK")
else:
  print(out)
```

```python
#DICTIONARY
#TASK-1
var1=input()
var2=input()
var1=var1.split(", ")
d1={}
for i in var1:
  i=i.split(": ")
  d1.update({i[0]:int(i[1])})
d2={}
var2=var2.split(", ")
for i in var2:
  i=i.split(": ")
  d2.update({i[0]:int(i[1])})
for key,value in d2.items():
  if key in d1.keys():
    d1[key]+=value
  else:
    d1[key]=value
print(d1)
out=[]
for i in d1.values():
  if i not in out:
    out.append(i)
print(tuple(sorted(out)))
#{'a': 400, 'b': 300, 'c': 200, 'd': 700, 'e': 200}
#(200, 300, 400, 700)
```

```python
#DICTIONARY
#TASK-1
var="{a: 100, b: 100, c: 200, d: 300},{a: 300, b: 200, d: 400, e: 200}"
var=var.split("},{")
var1,var2=var
var1=var1[1:len(var1):]
var1=var1.split(", ")
d1={}
for i in var1:
  i=i.split(": ")
  d1.update({i[0]:int(i[1])})
d2={}
var2=var2[:len(var2)-1:]
var2=var2.split(", ")
for i in var2:
  i=i.split(": ")
  d2.update({i[0]:int(i[1])})
for key,value in d2.items():
  if key in d1:
    d1[key]+=value
  else:
    d1[key]=value
print(d1)
out=[]
for i in d1.values():
  if i not in out:
    out.append(i)
print(tuple(sorted(out)))
```

```python
#TASk-4
dict1={1:(".",",",">","!",":"), 2:("A","B","C"), 3:("D","E","F"), 4:("G","H","I"), 5:("J","K","L"), 6:("M","N","O"), 7:("P","Q","R","S"), 8:("T","U","V"), 9:("W","X","Y","Z"), 0:(" ")}
var=input()
var=var.upper()
out=""
for char in var:
  for key,val in dict1.items():
    if char in val:
      for i in range(len(val)):
        if char==val[i]:
          num=i+1
          out+=str(key)*num
print(len(out))

```

```python
#SORTING
#ASCENDING
x = [5, 3, 2, 6, 4, 0]
n=len(x)
for i in range(n):
  for j in range(0,n-i-1):
    if x[j]>x[j+1]:
      x[j],x[j+1]=x[j+1],x[j]
print(x)
```

```python
#DESCENDING
x = [5, 4, 3, 2, 5, 1]
n = len(x)
for i in range(n):
  for j in range(0, n-i-1):
    if x[j+1] > x[j]:
        x[j], x[j+1] = x[j+1], x[j]
print(x)
```
