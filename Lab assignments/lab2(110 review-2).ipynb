```python
#TASK-1
def division(x,y):
  if y==0:
    return 0
  else:
    div=x/y-x//y
    if div>0:
      return f"{div:.1f}"
    else:
      return 0

print(division(10,3))
```

```python
#TASK-3
def func(start,end,divisor):
  sum=0
  for i in range(start,end):
    if i%divisor==0:
      sum+=i
  return sum
print(func(3,16,3))
```

```python
#TASK-4
def total(burger,location="Mohakhali"):
  sum=0
  if burger=="BBQ Chicken Cheese Burger":
    sum+=250*1.08
  elif burger=="Beef Burger":
    sum+=170*1.08
  elif burger=="Naga Drums":
    sum+=200*1.08
  if location!="Mohakhali":
    sum+=60
  else:
    sum+=40
  return f"{sum:.1f}"
print(total("Beef Burger","Dhanmondi"))
```

```python
#TASK-5
def replace_domain(user,new,old="kaaj.com"):
  for i in range(len(user)):
    if user[i]=="@":
      idx=i
      break
  name=user[0:idx]
  domain=user[idx+1:len(user)]
  out=""
  if domain==old:
    out+=name+"@"+"sheba.xyz"
    return f"Changed: {out}"
  elif domain==new:
    return f"Unchanged: {user}"
var=replace_domain('bob@sheba.xyz', 'sheba.xyz')
print(var)
```

```python
#TASK-6
def v_check(var):
  vowel="AEIOUaeiou"
  v=[]
  out=""
  for i in var:
    if i in vowel:
      v.append(i)
  for i in range(len(v)):
    if i==len(v)-1:
      out+=v[i]
    else:
      out+=v[i]+", "
  if len(v)==0:
    print("No vowels in the name")
  else:
    print(f"Vowels: {out}. Total number of vowels: {len(v)}")
v_check("Steve Jobs")
```

```python
#TASK-7
def palindrome(var):
  newvar=""
  for i in var:
    if i!=" ":
      newvar+=i
  palin=""
  for i in range(len(newvar)-1,-1,-1):
    palin+=newvar[i]
  if newvar==palin:
    print("palindrome")
  else:
    print("Not a Palindrome")
palindrome("nurses run")
```

```python
#TASK-8
def days(num):
  yy=num//365
  rem1=num%365
  mm=rem1//30
  dd=rem1%30
  return f"{yy} years, {mm} months and {dd} days"
print(days(4000))
```

```python
#TASK-9
def cap(text):
  temp=text[0].upper()
  i=1
  while i < len(text):
    if text[i] in ".?!" and i!=len(text)-1:
      temp+=text[i]+" "+ text[i+2].upper()
      i+=3
    elif text[i-1:i+2]==" i ":
      temp+="I"
      i+=1
    else:
      temp+=text[i]
      i+=1
  return temp
print(cap('my favourite animal is a dog. a dog has sharp teeth so that it can eat flesh very easily. do you know my pet dog’s name? so i love my pet very much.'))
```
